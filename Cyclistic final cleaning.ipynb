# CYCLISTIC BIKE SHARING DATA CLEANING

```python
# importing libraries
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
# reading in the file
bike_data = pd.read_csv(r'C:\Users\Gbolahan\Downloads\Samuel downloads\Divvy_trips\Divvy_Trips_2018_Q1.csv')
```

```python
bike_data.head()
```

```python
#  viewing the columns

bike_data.columns
```

```python
# renaming the columns

bike_data.columns = ['Rental_ID','Start_time','End_time','Bike_ID','Duration(secs)','Start_station_ID','Start_station_name','End_station_ID','End_station_name','User_type','Gender','DOB']
```

```python
bike_data.head()
```

```python
bike_data.info()
```

```python
# checking for nulls

bike_data.isnull().sum()
```

```python
# converting infinite values to nulls then droppung all nulls
bike_data.replace([np.inf, -np.inf], np.nan, inplace=True)
```

```python
# checking number of rows in datset before dropping nulls
len(bike_data)
```

```python
bike_data = bike_data.dropna()
```

```python
# checking number of rows in dataset after dropping nulls
len(bike_data)
```

```python
bike_data.describe()
```

```python
# cleaning the strings in the 'Duration (secs)' column and converting to appropriate datatype

bike_data['Duration(secs)'] = bike_data['Duration(secs)'].str.replace(',', '')

bike_data['Duration(secs)']= bike_data['Duration(secs)'].astype(float)

bike_data['Duration(secs)']= bike_data['Duration(secs)'].astype('timedelta64[s]')
```

```python
bike_data.info()
```

```python
# formatting column to datetime format
bike_data['Start_time']=pd.to_datetime(bike_data['Start_time'])
bike_data['End_time']=pd.to_datetime(bike_data['End_time'])
```

```python
# extracting year, month, day and day of the week
bike_data["year"]= bike_data["Start_time"].dt.year
bike_data["month"]= bike_data["Start_time"].dt.month
bike_data["day"]= bike_data["Start_time"].dt.day
bike_data["day_of_week"]= bike_data["Start_time"].dt.day_name()
```

```python
bike_data['DOB'] = pd.to_datetime(bike_data.DOB, format = '%Y')
bike_data["DOB"]=bike_data["DOB"].dt.year
```
